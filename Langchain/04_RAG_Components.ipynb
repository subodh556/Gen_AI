```python
#vector database

from langchain_community.document_loaders import TextLoader
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from langchain_chroma import Chroma

load = TextLoader("data/exx.txt").load()

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

chunk_of_text = text_splitter.split_documents(load)

vecdb = Chroma.from_documents(chunk_of_text, OpenAIEmbeddings())

question = "what did the president say about the economy?"

res = vecdb.similarity_search(question)

```

```python
#vector store as retriever
from langchain_community.vectorstores import FAISS
vecdbb = FAISS.from_documents(chunk_of_text, OpenAIEmbeddings())

retriever = vecdbb.as_retriever()

# if we want only 2 results then retriever = vecdbb.as_retriever(search_kwargs={"k":2})

print(retriever.invoke(question))
```
